```python
import pandas as pd
import numpy as np
import os
import json

def load_json(filename):
    """
    Load a JSON file given a filename
    If the file doesn't exist, then return an empty dictionary instead
    """
    try:
        with open(filename, 'r') as f:
            return json.load(f)
    except FileNotFoundError:
        return {}
```

```python
# output_dir = '../data/output/creative_short_v1_0/'

batch_id = 'creative_short_v1_0'
# batch_id = "creative_short_tmp_var"
exclude_lst = [
    'gemini_2_pro',
    '__MACOSX',
    'olmo_sft'
    
]

proj_dir = "/ix1/xli/bkb45/joey_files/creative_bench"

output_dir = '{}/{}/{}/'.format(proj_dir, "data/output", batch_id)
model_lst = [m for m in os.listdir(output_dir) if os.path.isdir(output_dir + m) and m not in exclude_lst]
model_lst
```

```python
result_dfs = []
for run_id in model_lst:
    tmp_df = pd.read_csv(output_dir + run_id + '/eval_report.csv')
    tmp_df['model'] = run_id
    result_dfs.append(tmp_df)
result_df = pd.concat(result_dfs)
result_df['avg_ngram'] = result_df.apply(
    lambda x: np.mean([
        x['0_gram'], x['1_gram']
    ]),
    axis = 1
)

result_df.sort_values(by = "avg_sur", ascending = False)
```
